```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

#### Curves to graph
- $ y = 0n + 1$ and label the curve "O(1)"
- $ y = log(n)$ and label the curve "O(log n)"
- $ y = n$ and label the curve "O(n)"
- $ y = n * log(n)$ and label it "O(n log n)"
- $ y = n^2$ and label it "O(n^2)"
- $ y = 2^n$ and label it "O(2^n)"
- $ y = n!$ and label it "O(n!)"
- $ y = n^n$ and label it "O(n^n)"

```python
# y = On + 1 and label the curve "O(1)"
x = np.linspace(0.01, 5, 400)
y1 = [0*n + 1 for n in x]

plt.figure(figsize=(8, 8))
plt.plot(x, y1, label='$y = 0n + 1$', ls='-', c='c')
plt.legend(loc='upper right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
#plt.xticks(list(plt.xticks()[0]) + list(range(-4,4, 1)))
#plt.yticks(list(plt.yticks()[0]) + list(np.arange(-50, 50, 50)))
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = log(n) and label the curve "O(log n)"
x = np.linspace(0.01, 5, 400)
y2 = np.log(x)

plt.figure(figsize=(8, 8))
plt.plot(x, y2, label='$y=O(log n)$', ls='-', c='red')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = n and label the curve "O(n)"
x = np.linspace(0.01, 5, 400)
y3 = x

plt.figure(figsize=(8, 8))
plt.plot(x, y3, label='$y=O(n)$', ls='-', c='purple')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = n * log(n) and label it "O(n log n)"
x = np.linspace(0.01, 5, 400)
y4 = x * np.log(x)

plt.figure(figsize=(8, 8))
plt.plot(x, y4, label='$y=O(n*\log(n))$', ls='-', c='maroon')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = n^2 and label it "O(n^2)"
x = np.linspace(0.01, 5, 400)
y5 = x**2

plt.figure(figsize=(8, 8))
plt.plot(x, y5, label='$y=O(n^2)$', ls='-', c='m')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = 2^n and label it "O(2^n)"
x = np.linspace(0.01, 5, 400)
y6 = 2**x

plt.figure(figsize=(8, 8))
plt.plot(x, y6, label='$y=O(2^n)$', ls='-', c='green')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
import math
# y = n! and label it "O(n!)"
x2 = list(range(0,5))
y7 = [math.factorial(i) for i in x2]

plt.figure(figsize=(8, 8))
plt.plot(x2, y7, label='$y=O(n!)$', ls='-', c='blue')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.xticks(list(plt.xticks()[0]) + list(range(0,5)))
plt.yticks(list(plt.yticks()[0]) + list(np.arange(-5, 30)))
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```

```python
# y = n^n and label it "O(n^n)"
x = np.linspace(0.01, , 400)
y8 = (x**x)

plt.figure(figsize=(8, 8))
plt.plot(x, y8, label='$y=O(n^n)$', ls='-', c='blue')
plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle=':', linewidth=0.5, c='lightgray')
plt.title('Big O Notation', fontsize=24)
plt.xlabel('Elements', fontsize=16)
plt.ylabel('Operations', fontsize=16)
plt.show()
```
