#### 1. Use matplotlib to plot the following equation:

$y =x^{2} - x + 2$

##### You'll need to write the code that generates the x and y points.
##### Add an anotation for the point 0, 0, as the origin.

```python
x = np.linspace(0, 8, 400)
y = [n**2 - n + 2 for n in x]

plt.figure(figsize=(16, 9))
plt.plot(x, y, c='r')
plt.xlabel('$x$')
plt.ylabel('$x^2-x+2$')
plt.title('$y=x^2-x+2$ Line ')
plt.annotate('Origin', xy=(0,0), xytext=(0,10),arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, c='lightgray')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
x = range(-7, 8)
y = [n**2 - n + 2 for n in x]

# Create the figure and axis with a specified size
plt.figure(figsize=(16, 9))

# Plot the data with a colored line (e.g., blue)
plt.plot(x, y, color='cyan', linewidth=2.5, label='$y=x^2-x+2$')

# Set labels with larger font sizes
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y = x^2 - x + 2$', fontsize=14)
plt.title('$y=x^2-x+2$ Line Plot', fontsize=24, c='k')

# Annotate the origin
plt.annotate('Origin', xy=(0, 0), xytext=(-2.8, 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12)

# Display grid lines
plt.grid(True, which='both', linestyle='--', linewidth=0.5, c='lightgray')

# Set a light grey color for the zero lines of x and y to distinguish them from the grid
plt.axhline(0, color='grey', linewidth=0.8)
plt.axvline(0, color='grey', linewidth=0.8)

# Add ticks for better precision
plt.xticks(list(plt.xticks()[0]) + [-7,-6,-5,-4,-3,-2,-1,1,2,3,4,5,6,7,8])
plt.yticks(list(plt.yticks()[0]) + list(range(0, 65, 5)))

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_letter_J(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 3), 2, 0.5, facecolor="blue"))
    ax.add_patch(patches.Rectangle((x_offset+1.5, 0.5), 0.5, 3, facecolor="blue"))
    ax.add_patch(patches.Rectangle((x_offset, 0), 1.5, 0.5, facecolor="blue"))

def plot_letter_o(ax, x_offset):
    ax.add_patch(patches.Ellipse((x_offset+1, 2), width=2, height=3, facecolor="red"))

def plot_letter_r(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 2.5), 0.5, 1.5, facecolor="green"))
    ax.add_patch(patches.Rectangle((x_offset, 3.5), 1.5, 0.5, facecolor="green"))
    ax.add_patch(patches.PathPatch(patches.Path([(x_offset, 2.5), (x_offset+1.5, 2.5), (x_offset, 1)], closed=False), facecolor="none", edgecolor="green", lw=0.5))

def plot_letter_g(ax, x_offset):
    ax.add_patch(patches.Ellipse((x_offset+1, 1.5), width=2, height=2.5, facecolor="purple"))
    ax.add_patch(patches.Rectangle((x_offset, 0), 0.5, 1, facecolor="purple"))

def plot_letter_e(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 3), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 2), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 1), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 1), 0.5, 2, facecolor="orange"))

# Create the plot
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_xlim(0, 18)
ax.set_ylim(0, 5)
ax.set_aspect('equal', 'box')
ax.axis('off')

# Plot each letter
plot_letter_J(ax, 0)
plot_letter_o(ax, 3)
plot_letter_r(ax, 6)
plot_letter_g(ax, 9)
plot_letter_e(ax, 12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_letter_J(ax, x_offset):
    # Top horizontal
    ax.add_patch(patches.Rectangle((x_offset, 4), 2, 0.5, facecolor="blue"))
    # Vertical right
    ax.add_patch(patches.Rectangle((x_offset+1.5, 1.5), 0.5, 3, facecolor="blue"))
    # Bottom curve
    ax.add_patch(patches.Rectangle((x_offset, 1), 1.5, 0.5, facecolor="blue"))
    ax.add_patch(patches.Rectangle((x_offset, 1.5), 0.5, 1, facecolor="blue"))

def plot_letter_O(ax, x_offset):
    # Top horizontal
    ax.add_patch(patches.Rectangle((x_offset, 4), 3, 0.5, facecolor="red"))
    # Bottom horizontal
    ax.add_patch(patches.Rectangle((x_offset, 0), 3, 0.5, facecolor="red"))
    # Left vertical
    ax.add_patch(patches.Rectangle((x_offset, 0), 0.5, 4.5, facecolor="red"))
    # Right vertical
    ax.add_patch(patches.Rectangle((x_offset+2.5, 0), 0.5, 4.5, facecolor="red"))

def plot_letter_R(ax, x_offset):
    # Top horizontal
    ax.add_patch(patches.Rectangle((x_offset, 4), 3, 0.5, facecolor="green"))
    # Left vertical
    ax.add_patch(patches.Rectangle((x_offset, 0), 0.5, 4.5, facecolor="green"))
    # Middle horizontal
    ax.add_patch(patches.Rectangle((x_offset, 2.25), 2.5, 0.5, facecolor="green"))
    # Diagonal
    for i in range(8):
        ax.add_patch(patches.Rectangle((x_offset + 0.25*(i+1), 2 - 0.25*i), 0.5, 0.5, facecolor="green"))

def plot_letter_G(ax, x_offset):
    # Top horizontal
    ax.add_patch(patches.Rectangle((x_offset, 4), 3, 0.5, facecolor="purple"))
    # Bottom horizontal
    ax.add_patch(patches.Rectangle((x_offset, 0), 3, 0.5, facecolor="purple"))
    # Left vertical
    ax.add_patch(patches.Rectangle((x_offset, 0), 0.5, 4.5, facecolor="purple"))
    # Middle horizontal
    ax.add_patch(patches.Rectangle((x_offset, 2.25), 2, 0.5, facecolor="purple"))

def plot_letter_E(ax, x_offset):
    # Top horizontal
    ax.add_patch(patches.Rectangle((x_offset, 4), 2.5, 0.5, facecolor="orange"))
    # Middle horizontal
    ax.add_patch(patches.Rectangle((x_offset, 2.25), 2.5, 0.5, facecolor="orange"))
    # Bottom horizontal
    ax.add_patch(patches.Rectangle((x_offset, 0), 2.5, 0.5, facecolor="orange"))
    # Left vertical
    ax.add_patch(patches.Rectangle((x_offset, 0), 0.5, 4.5, facecolor="orange"))

# Create the plot
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_xlim(0, 18)
ax.set_ylim(0, 5)
ax.set_aspect('equal', 'box')
ax.axis('off')

# Plot each letter
plot_letter_J(ax, 0)
plot_letter_O(ax, 3)
plot_letter_R(ax, 7)
plot_letter_G(ax, 11)
plot_letter_E(ax, 15)

plt.show()
```

#### 7. Bonus: use curves for letters in your name that have curves in them.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_letter_J(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 3), 2, 0.5, facecolor="blue"))
    ax.add_patch(patches.Rectangle((x_offset+1.5, 0.5), 0.5, 3, facecolor="blue"))
    ax.add_patch(patches.Arc((x_offset+1, 0.5), 2, 1, theta1=0, theta2=180, facecolor="blue", edgecolor="blue", lw=0.5))

def plot_letter_o(ax, x_offset):
    ax.add_patch(patches.Ellipse((x_offset+1, 2), width=2, height=3, facecolor="red"))

def plot_letter_r(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 2.5), 0.5, 1.5, facecolor="green"))
    ax.add_patch(patches.Rectangle((x_offset, 3.5), 1.5, 0.5, facecolor="green"))
    ax.add_patch(patches.PathPatch(patches.Path([(x_offset, 2.5), (x_offset+1.5, 2.5), (x_offset, 1)], closed=False), facecolor="none", edgecolor="green", lw=0.5))

def plot_letter_g(ax, x_offset):
    circle_top = patches.Ellipse((x_offset+1, 2), width=2, height=2.5, facecolor="purple")
    arc_bottom = patches.Arc((x_offset+1, 0.5), 2, 1, theta1=0, theta2=180, facecolor="purple", edgecolor="purple", lw=4)
    ax.add_patch(circle_top)
    ax.add_patch(arc_bottom)

def plot_letter_e(ax, x_offset):
    ax.add_patch(patches.Rectangle((x_offset, 3), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 2), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 1), 1.5, 0.5, facecolor="orange"))
    ax.add_patch(patches.Rectangle((x_offset, 1), 0.5, 2, facecolor="orange"))

# Create the plot
fig, ax = plt.subplots(figsize=(16, 6))
ax.set_xlim(0, 18)
ax.set_ylim(0, 5)
ax.set_aspect('equal', 'box')
ax.axis('off')

# Plot each letter
plot_letter_J(ax, 0)
plot_letter_o(ax, 3)
plot_letter_r(ax, 6)
plot_letter_g(ax, 9)
plot_letter_e(ax, 12)

plt.show()
```
